# Token
Token作为AI模型中用于计算文本长度的基本单位。
他们是一组字符，token并不总是和单词对应。

## Tiktoken
使用开源工具快速分割token

```python
pip install tiktoken
```

```python
import openai
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4")

en = """We should create our own future."""

tokens = encoding.encode(en)

print(tokens)

num_of_tokens_in_en = len(encoding.encode(en))

print(f"chinese:{en} ; {num_of_tokens_in_en} tokens\n")
```
